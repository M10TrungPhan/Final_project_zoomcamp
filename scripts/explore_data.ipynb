```python
list_df_station_start = []
list_df_station_end = []
folder_data = '../data/'
for year in [2022, 2023]:
    for month in range(1,13):
        month = "{:02}".format(month)
        name_file = f"{str(year)}{month}-divvy-tripdata.csv"
        if name_file  in os.listdir(folder_data):
            df = pd.read_csv(f'{folder_data}{year}{month}-divvy-tripdata.csv', dtype=str)
            df_start = df.groupby(['start_station_name','start_station_id']).size().reset_index(name='counts')
            df_end =  df.groupby(['end_station_name','end_station_id']).size().reset_index(name='counts')
            list_df_station_start.append(df_start)
            list_df_station_end.append(df_end)
```

```python
pd.concat(list_df_station_start).groupby(['start_station_name','start_station_id']).sum().reset_index()
```

```python
pd.concat(list_df_station_end).groupby(['end_station_name','end_station_id']).sum().reset_index()
```

```python
list_id_2022 = []
for each in  df['start_station_id'].unique():
    each = str(each)
    if re.search(r"^\d+\.0$",each) is not None:
        each = re.sub(r'\.0$', '', each).strip()
        list_id_2022.append(each)
    else:
        list_id_2022.append(each)
```

```python
list_id_crawl = df2['Short Name'].unique() 
id_not_match = []
for each in list_id_2022:
    if each not in list_id_crawl:
        id_not_match.append(each)
```
